```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
import seaborn as sns
import shutil
```

```python
# Read CSV from dataset
diagnosis = pd.read_csv('CoronaHack-Respiratory-Sound-Dataset/Corona-Hack-Respiratory-Sound-Metadata.csv', encoding='cp1252')
diagnosis.head()
```

```python
#read d
cough_heavy = [x for x in diagnosis['cough-heavy'] if pd.isnull(x) == False]
cough_shallow = [x for x in diagnosis['cough-shallow'] if pd.isnull(x) == False]
counting_fast = [x for x in diagnosis['counting-fast'] if pd.isnull(x) == False]
counting_normal = [x for x in diagnosis['counting-normal'] if pd.isnull(x) == False]
vowel_a = [x for x in diagnosis['vowel-a'] if pd.isnull(x) == False]
vowel_e = [x for x in diagnosis['vowel-e'] if pd.isnull(x) == False]
vowel_o = [x for x in diagnosis['vowel-o'] if pd.isnull(x) == False]
```

```python
#create a directory for the result of this process
os.mkdir('dataset')
os.mkdir('dataset/cough_heavy')
os.mkdir('dataset/cough_shallow')
os.mkdir('dataset/talk')
```

```python
a = 0
for f in cough_heavy:
    original = 'CoronaHack-Respiratory-Sound-Dataset'+f
    target = 'dataset/cough_heavy/cough-heavy-'+str(a)+'.wav'
    shutil.copyfile(original, target)
    print(target)
    a = a+1 
```
